```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences =["The book on the shelf",
"A cat is sleeping on the windowsill",
"She played a beautiful melody on the piano",
"The sun sets behind the mountains",
"Learning new languages opens up new opportunities",
"Exploring different cultures enhances our understanding of the world",
"Coding is a valuable skill in today's technology-driven society",
"The smell of fresh coffee in the morning",
"Running along the beach at sunset",
"Solving complex puzzles sharpens the mind"]
```

```python
vocabulary_size =10000
one_hot_repr = [one_hot(words,vocabulary_size)for words in sentences]
print(one_hot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
sent_length = 10
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim = 10
model = Sequential()
model.add(Embedding(vocabulary_size,10,input_length=sent_length))
model.summary()
```

```python
print(model.predict(embedded_docs))
```
